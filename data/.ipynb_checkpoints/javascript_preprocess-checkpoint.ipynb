## 1. Process data

### 1.1 Load datasets

```python
# Trees csv

df = pd.read_csv("./PPR_Tree_Inventory_2022.csv")
df.head()
```

```python
# Load trees

gpd_df = gpd.read_file("./PPR_Tree_Inventory_2022.geojson")
gpd_df.head()
```

```python
# Load neighbourhoods

neighborhoods = gpd.read_file(
    "https://raw.githubusercontent.com/azavea/geo-data/master/Neighborhoods_Philadelphia/Neighborhoods_Philadelphia.geojson"
)
neighborhoods.head()
```

```python
# Load census tracts

tract = gpd.read_file("https://opendata.arcgis.com/datasets/8bc0786524a4486bb3cf0f9862ad0fbf_0.geojson")
tract.head()
```

### 1.2 Spatially join tree to a neighbourhood and census tract

```python
# Join tree to tract

join_1 = gpd.sjoin(
    gpd_df,
    tract[['geometry','GEOID10', 'NAME10']], 
    predicate="within",
    how="left",
)
join_1 = join_1.drop(['index_right'], axis=1)
join_1.head()
```

```python
# Join to neighbourhood

join_2 = gpd.sjoin( 
    join_1,
    neighborhoods.to_crs(join_1.crs), # REMEMBER: convert to the right CRS
    predicate="within",
).reset_index(drop=True)

join_2 = join_2.drop(['name', 'created_at', 'updated_at', 'listname'], axis=1)
join_2.head()
```

### 1.3 Explore data

#### 1.3.1 Frequency of trees

```python
# Group by tree & find top tree frequency

grp_tree = join_2.groupby("TREE_NAME").size().reset_index(name='size')
grp_tree.sort_values('size', ascending=False).head(10)
```

```python
# Merge neighbourhood geometries back in
neighborhoods2 = neighborhoods[['mapname', 'geometry']]
grp_nh2 = neighborhoods2.merge(grp_nh, on="mapname")
grp_nh2 = grp_nh2.to_crs(epsg=3857)
```
